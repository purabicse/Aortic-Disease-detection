```python
import imageio
import os
from PIL import Image

def robust_extract(video_path, output_folder):
    if not os.path.exists(output_folder):
        os.makedirs(output_folder)

    try:
        # 'ffmpeg' plugin is much better at reading converted .mp4 headers
        reader = imageio.get_reader(video_path, 'ffmpeg')

        count = 0
        for frame in reader:
            # frame is a numpy array (RGB)
            img = Image.fromarray(frame)
            img.save(os.path.join(output_folder, f"frame_{count:05d}.jpg"))
            count += 1

        print(f"Success! Extracted {count} frames from {video_path}")

    except Exception as e:
        print(f"Could not extract {video_path}. Error: {e}")

# Try it on one of your 18 converted samples
robust_extract("/content/36.mp4", "/content/drive/MyDrive/PSAX aortic frames/p36")
```
